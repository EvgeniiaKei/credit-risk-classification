```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report, balanced_accuracy_score
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
df_lending = pd.read_csv(
    Path('Resources/lending_data.csv')
)
# Review the DataFrame
print(df_lending.shape)
df_lending.head()
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = df_lending["loan_status"]

# Separate the X variable, the features
X = df_lending.drop("loan_status", axis=1)
```

```python
# Review the y variable Series
y[:10]
```

```python
# Review the X variable DataFrame
X[:10]
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
classifier = LogisticRegression(random_state=1, solver='lbfgs', max_iter=200)

# Fit the model using training data
classifier.fit(X_train, y_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).tail(10)
```

### Step 3: Evaluate the model’s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# Generate a confusion matrix for the model
confuse_matrix = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(confuse_matrix,
                      index=["Actual 0", "Actual 1"],
                      columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Print the classification report for the model
target_names = ["(0) Healthy Loans","(1) High-Risk Loans"]
print(classification_report(y_test,predictions,target_names=target_names))
```

### Step 4: Answer the following question.

**Question:** How well does the logistic regression model predict both the `0` (healthy loan) and `1` (high-risk loan) labels?

**Answer:** The model has an impressive overall accuracy of 99%, indicating that it correctly classifies the vast majority of instances. However, there is a gap in the models ability to predict high-risk loans (1), with about 10% of high-risk loans being classified as healthy. This is possibly due to having significantly less data points of high-risk loans to use for training.
The macro average metrics (precision: 0.92, recall: 0.97, F1-score: 0.94) indicate good overall performance, treating both classes equally.
The weighted averages (precision: 0.99, recall: 0.99, F1-score: 0.99) reflect the model's performance considering the class distribution, showing that it performs exceptionally well overall, particularly for healthy loans.
In summary, the model is strong in predicting healthy loans and reasonably good in predicting high-risk loans, but there’s potential for improvement in the latter.
